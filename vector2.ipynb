```python
nlp = spacy.load("ru_core_news_sm")
```

```python
text = '''У лукоморья дуб зелёный;
Златая цепь на дубе том:
И днём и ночью кот учёный
Всё ходит по цепи кругом;
Идёт направо — песнь заводит,
Налево — сказку говорит.
Там чудеса: там леший бродит,
Русалка на ветвях сидит;
Там на неведомых дорожках
Следы невиданных зверей;
Избушка там на курьих ножках
Стоит без окон, без дверей;
Там лес и дол видений полны;
Там о заре прихлынут волны
На брег песчаный и пустой,
И тридцать витязей прекрасных
Чредой из вод выходят ясных,
И с ними дядька их морской;
Там королевич мимоходом
Пленяет грозного царя;
Там в облаках перед народом
Через леса, через моря
Колдун несёт богатыря;
В темнице там царевна тужит,
А бурый волк ей верно служит;
Там ступа с Бабою Ягой
Идёт, бредёт сама собой,
Там царь Кащей над златом чахнет;
Там русский дух… там Русью пахнет!
И там я был, и мёд я пил;
У моря видел дуб зелёный;
Под ним сидел, и кот учёный
Свои мне сказки говорил.'''
```

```python
nlpText = nlp(text)
```

```python
print(nlpText)
```

```python
Stop = nlp.Defaults.stop_words
```

```python
StopList = sorted(list(Stop))
```

```python
for StopWord in StopList:
    print(StopWord)
```

```python
FilteredWords = [token.text for token in nlpText if not token.is_stop]
filteredText = ' '.join(FilteredWords)
print(filteredText)
```

```python
nlpFilteredText = nlp(filteredText)
VectorText = nlpText.vector
VectorFilteredText = nlpFilteredText.vector
```

```python
print(VectorText)
```

```python
print(VectorFilteredText)
```

```python
distance = cosine(VectorText,VectorFilteredText)
print(distance)
similar = nlpText.similarity(nlpFilteredText)
print(similar)
cosinDis = 1 - similar
print(cosinDis)
```

```python
Sentences = list(nlpText.sents)
```

```python
SentenceVectors = {}
for i,Sentence in enumerate(Sentences):
    SentenceText = nlp(Sentence.text)
    filteredSentenceText = ' '.join([token.text for token in SentenceText if not token.is_stop])
    print(filteredSentenceText)
    filteredSentenceVector = nlp(filteredSentenceText).vector
    SentenceVectors[f"sentence_{i+1}"] = filteredSentenceVector
    print(filteredSentenceVector)
```
